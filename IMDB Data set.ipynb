```python
#importing libraries pandas for data frame reading and data anlaysis,matlplot,seaborn for visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#READING A DATA FRAME USING PANDAS
df=pd.read_csv("S:\\Datascience\\IMDB-Movie-Data.csv")
```

```python
#display first 5 values use head
df.head()
```

```python
#to get last 5 values use tail()
df.tail()
```

```python
#to get information about data set
df.info()
```

```python
#checking missing values
df.isnull().sum()
```

```python
#plotting data using heat map for null values
sns.heatmap(df.isnull())
plt.show()
```

```python
#to get missing value percentage
percentage_missing=df.isnull().sum()*100/len(df)
print(percentage_missing)
```

```python
#drop all missing values
df.dropna(axis=0)
```

```python
#check for duplicate data
duplicate_data=df.duplicated().any()
print(duplicate_data)
```

```python
#if there is duplicte we have to follow below line
df=df.drop_duplicates()
```

```python
#to get statistics about dataframe if we use include all then it get NaN values 
df.describe(include='all')
```

```python
df.describe()
```

```python
df.columns
```

```python
#display title of  the movie having runtime >=200minutes
df[df['Runtime (Minutes)']>=160]['Title']
```

```python
#in which year there was the highest average voting
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
#data visualization
sns.barplot(x='Year',y='Votes',data=df)
plt.title("Average votes by year")
plt.show()
```

```python
#In which year there was the highest average revenue
df.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=df)
plt.title("Average votes by year")
plt.show()
```

```python
#to get the average rating for each director
df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
#to display top 10 length movies title and runtime
top10=df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]\
.set_index('Title')
print(top10)
```

```python
sns.barplot(x='Runtime (Minutes)',y=top10.index,data=top10)
plt.show()
```

```python
#to get no of movies per year
df['Year'].value_counts()
```

```python
sns.countplot(x='Year',data=df)
plt.show()
```

```python
#to print most popular movie title(Highest revenue)
df[df['Revenue (Millions)'].max()==df['Revenue (Millions)']]['Title']
```
